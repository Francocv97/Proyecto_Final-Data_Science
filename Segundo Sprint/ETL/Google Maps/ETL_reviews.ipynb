```python
import pandas as pd
import json

def procesar_archivo_json(ruta_archivo):
    data = []
    with open(ruta_archivo, 'r') as file:
        for line in file:
            try:
                obj = json.loads(line)
                data.append(obj)
            except json.JSONDecodeError as e:
                continue

    df = pd.DataFrame(data)
    df['time'] = pd.to_datetime(df['time'], unit='ms')
    df = df[['user_id', 'name', 'time', 'rating', 'text', 'gmap_id']].dropna(subset=['text']).reset_index(drop=True)
    return df

# Lista de rutas de los archivos JSON
rutas_json = ['/content/PA_20porciento.json', '/content/IL_20porciento.json', '/content/FL_20porciento.json', '/content/CA_20porciento.json']

# Procesar cada archivo JSON y guardar los resultados en una lista
dfs_procesados = []
for ruta_json in rutas_json:
    df_procesado = procesar_archivo_json(ruta_json)
    dfs_procesados.append(df_procesado)

# Imprimir los dataframes procesados
for i, df_procesado in enumerate(dfs_procesados):
    print(f"\n--------------------------------------------------\nDataFrame {i+1} procesado:\n")
    display(df_procesado)
```
