```python
# Guarda las rutas de los data sets  en variables
ruta1='/content/drive/MyDrive/Proyecto Final/Yelp/business.pkl'
ruta2='/content/drive/MyDrive/Proyecto Final/Yelp/checkin.json'
ruta3='/content/drive/MyDrive/Proyecto Final/Yelp/review.json'
ruta4='/content/drive/MyDrive/Proyecto Final/Yelp/tip.json'
ruta5='/content/drive/MyDrive/Proyecto Final/Yelp/user.parquet'
```

```python
print('Dataset "business" de la plataforma Yelp:\n')
business=pd.read_pickle(ruta1)
parte_nula=business.iloc[:,14:].tail()
parte_nula.dropna(how='all')
business=business.iloc[:,:14]
business
```

### **Eliminación de columnas del conjunto de datos**
---
Las columnas "is_open" , "attributes" y "hours" serán eliminadas del conjunto de datos, dicha decisión fue tomada durante el análisis exploratorio de los datos (EDA) dado que las dimensiones mencionadas no aportan información reelevante al objetivo del proyecto.

```python
# Elimina columnas innecesarias
df_business=business.drop(columns=['is_open','attributes','hours'])
df_business
```

```python
df_business.info()
```

### **Valores Nulos**
---
Se decide eliminar los registros con valores nulos del conjunto de datos, ya que la cantidad es insignificante respecto a la totalidad de datos.

```python
df_business.isnull().sum()
```

```python
df_business=df_business.dropna().reset_index(drop=True)
df_business.isnull().sum()
```

### **Cambio de Tipos**
---
Las columnas "latitude" , "longitude" , "stars" , "review_count" contienen datos de tipo numerico, las mismas están en el tipo incorrecto , por lo cual se procederá a corregirlo.


```python
# Cambio de Tipos
df_business['latitude']=df_business['latitude'].astype('float')
df_business['longitude']=df_business['longitude'].astype('float')
df_business['stars']=df_business['stars'].astype('int')
df_business['review_count']=df_business['review_count'].astype('int')
df_business.info()
```

### Filtado de Estados
---
Del análisis exploratorio se determinó que no trabajariamos con la todalidad de lso estados sino con los 4 estados con mas negocios por región,estos fueron Pensilvania (PA), Florida (FL), California (CA) e Ilinois (IL), los mismos fueron seleccionados en base a los datos de la plataforma "Yelp" en conjunto con los datos de "Google Maps"


```python
# Lista de estados seleccionados
estados_seleccionados=['PA','FL','CA','IL']
# Filtado de estados
df_business=df_business[df_business['state'].isin(estados_seleccionados)].reset_index(drop=True)
df_business
```

```python
maximo=df_business['review_count'].max()
minimo=df_business['review_count'].min()
print(f'Valor maximo {maximo}, valor minimo {minimo}')
```
