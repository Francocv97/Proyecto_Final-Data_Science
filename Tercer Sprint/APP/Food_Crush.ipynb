```python
# Creamos la app
%%writefile Food_Crush.py

# Importamos librerías:
import streamlit as st
import lottie
import time
import streamlit_option_menu
from streamlit_option_menu import option_menu
import emoji

# Posición correcta:
st.set_page_config(
    page_title="Food Crush!🍔",
    layout="wide",
    initial_sidebar_state="expanded",
)

# Inicializar selected_option fuera del bloque with
selected_option = None

# Crea la barra lateral
with st.sidebar:
    # Menu options
    selected_option = option_menu(
        menu_title="Menú principal:",
        options=["Inicio", "Estados", "Recomendaciones", "Restaurantes", "Acerca de", "Contacto"],
        default_index=0,  # Set default option to "Inicio" (index 0)
        icons=["cast", "map", "book", "restaurant", "info", "envelope"],
        menu_icon="bars",
    )

# Título e ícono
st.title("Food Crush! " + emoji.emojize(":hamburger:"))

def main():
    # Contenido de la página
    if selected_option == "Inicio":
        st.write("¡Hola! ¿Todavía no comiste? ¿Qué esperas para aprovechar los descuentos? :sunglasses:")
        st.write("Aquí puedes encontrar información sobre los mejores restaurantes, bares y consejos para una experiencia inolvidable.")
        st.write("**Explora nuestras opciones:**")
        st.write("- **Recomendaciones:** Encuentra tu lugar ideal para deleitarte con comidas y tragos.")
        st.write("- **Restaurantes:** Descubre nuevos lugares para ir a comer :heart_eyes:.")
        st.write("- **Consejos:** Mejora tus lugares preferidos :wink:.")
        st.write("- **Descuentos:** Hasta un 60% OFF en restaurantes y beneficios exclusivos, para usuarios de Food Crush! :star_struck:")

    elif selected_option == "Recomendaciones":
        st.write("Buscador de recomendaciones:")

        # Botones de selección de estado
        state_button = st.button("Seleccionar estado")
        selected_state = None

        if state_button:
            selected_state = st.selectbox("Seleccione un estado", ["Pensilvania", "Texas"])

            if selected_state:
                st.write(f"Estado seleccionado: {selected_state}")

    # ... Resto del código para las demás opciones ...

if __name__ == "__main__":
    main()
```
