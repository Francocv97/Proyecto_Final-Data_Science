```python
# Instalamos librerías
import pandas as pd
import nltk
import swifter
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
# Creamos la instancia para el analizador de sentimientos
sia = SentimentIntensityAnalyzer()
```

```python
# Creamos la función para analizar el sentimiento:
def get_sentiment_label(compound_score):
    if compound_score >= 0.05:
        return 2  # Positivo
    elif compound_score <= -0.05:
        return 0  # Negativo
    else:
        return 1  # Neutro
```

```python
# Definimos una función para obtener el sentimiento:
# Función para obtener el sentimiento y etiquetas de un texto
def get_sentiment_info(text):
    sentiment_scores = sia.polarity_scores(text)
    sentiment_label = get_sentiment_label(sentiment_scores['compound'])
    return sentiment_scores['compound'], sentiment_label
```

```python
# Muestreo aleatorio del 20% del DataFrame original
#df_muestra = df_cuatroestados_unico.sample(frac=0.2, random_state=42)
```

```python
# Filtrar por estado
estado_deseado = 'PA'
df_estado = df_cuatroestados_unico[df_cuatroestados_unico['estado'] == estado_deseado].copy()

# Comprobación de valores nulos en la columna 'text'
df_estado = df_estado.dropna(subset=['text'])

# Aplicar la función de análisis de sentimiento a la columna 'text' de manera paralela en el DataFrame filtrado
sentiment_info = df_estado['text'].swifter.apply(lambda text: get_sentiment_info(text) if pd.notnull(text) else (0, ''))

# Asignar los resultados a nuevas columnas
df_estado = df_estado.assign(sentiment_scores=sentiment_info.apply(lambda x: x[0]),
                             sentiment_label=sentiment_info.apply(lambda x: x[1]))

# Ordenar por puntaje de sentimiento en orden descendente para obtener las recomendaciones
df_estado_recomendaciones = df_estado.sort_values(by='sentiment_label', ascending=False)
```

```python
# Mostrar las recomendaciones
print(f"Recomendaciones para el estado {estado_deseado}:\n", df_estado_recomendaciones[['restaurant_name','text', 'sentiment_scores', 'sentiment_label']])
```

```python
from IPython.display import display

# Mostrar solo la columna de análisis de sentimiento
display(df_estado_recomendaciones[['restaurant_name','text', 'sentiment_label']])
```

```python
# Función del top 5 de los restaurantes más recomendados por cada estado:


def analizar_sentimiento_y_top5(df, estado):
    # Filtrar por estado
    df_estado = df[df['estado'] == estado].copy()

    # Verificar si hay datos para el estado dado
    if df_estado.empty:
        print(f'No hay datos para el estado {estado}.')
        return

    # Aplicar la función de análisis de sentimiento a la columna 'text' de manera paralela en el DataFrame filtrado
    sentiment_info = df_estado['text'].swifter.apply(lambda text: get_sentiment_info(text))

    # Asignar los resultados a nuevas columnas
    df_estado = df_estado.assign(sentiment_scores=sentiment_info.apply(lambda x: x[0]),
                                sentiment_label=sentiment_info.apply(lambda x: x[1]))

    # Ordenar por puntaje de sentimiento en orden descendente para obtener las recomendaciones
    df_estado_recomendaciones = df_estado.sort_values(by='sentiment_scores', ascending=False)

    # Mostrar solo la columna de análisis de sentimiento
    print(f'\nAnalisis de sentimiento para el estado {estado}:\n')
    display(df_estado_recomendaciones[['restaurant_name', 'text', 'sentiment_label']])

    # Top 5 de restaurantes más frecuentes
    top5_restaurantes = df_estado['restaurant_name'].value_counts().nlargest(5)
    if top5_restaurantes.empty:
        print(f'\nNo hay restaurantes frecuentes para el estado {estado}.\n')
    else:
        print(f'\nTop 5 de restaurantes más frecuentes para el estado {estado}:\n')
        print(top5_restaurantes)

# Ejemplo de uso
analizar_sentimiento_y_top5(df_cuatroestados_unico, 'CA')
```

```python
# Ejemplo de uso
analizar_sentimiento_y_top5(df_cuatroestados_unico, 'FL')
```
